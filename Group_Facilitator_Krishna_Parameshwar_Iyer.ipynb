## Bivariant Analysis

```python
#Bivariant Analysis

Income_vs_lone_status = df1.groupby(['Income_amnt_range', 'loan_status']).loan_status.count().unstack().reset_index()
```

```python
#Calculating the Charged off Ratio from the total 

Income_vs_lone_status['Total_mem'] = Income_vs_lone_status['Charged Off'] + Income_vs_lone_status['Current'] + Income_vs_lone_status['Fully Paid']
Income_vs_lone_status
Income_vs_lone_status['Charged_off_Ratio'] = Income_vs_lone_status['Charged Off']/Income_vs_lone_status['Total_mem']
```

```python
Income_vs_lone_status
```

```python
#Plot for Annual Income vs Charged off Ratio 

sns.barplot(x= 'Charged_off_Ratio', y = 'Income_amnt_range', data = Income_vs_lone_status)
plt.title('Annual Income Amount vs Charged off Ratio', fontsize = 14)
plt.xlabel("Charged off Ratio", fontsize = 12)
plt.ylabel("Annual Income Amount Range", fontsize = 12)

#Observation 
#Those for whome the Annual Income range fall between 0-20000 fall more under charged off . 
# Borrowers with low annual income generally are Charged off 
```

```python
loan_amnt_vs_lone_status = df1.groupby(['Loan_amnt_range', 'loan_status']).loan_status.count().unstack().reset_index()
loan_amnt_vs_lone_status['Total_mem'] = loan_amnt_vs_lone_status['Charged Off'] + loan_amnt_vs_lone_status['Current'] + loan_amnt_vs_lone_status['Fully Paid']
loan_amnt_vs_lone_status
loan_amnt_vs_lone_status['Charged_off_Ratio'] = loan_amnt_vs_lone_status['Charged Off']/loan_amnt_vs_lone_status['Total_mem']
```

```python
loan_amnt_vs_lone_status
```

```python
#Plot for Loan Amount vs Charged off Ratio 

sns.barplot(y= 'Charged_off_Ratio', x = 'Loan_amnt_range', data = loan_amnt_vs_lone_status)
plt.title('Loan Amount vs Charged off Ratio', fontsize = 14)
plt.xlabel("Loan Amount Range", fontsize = 12)

#Observation
# Chargedoff Ratio is high for the Borrowers taken loan amount more than 28000
```
